
#2. Take input A and B and print their sum, mul, div, square of the nos.
##Ans:
```python
A = int(input("Enter a number A: "))
B = int(input("Enter a number B: "))
print(type(A))
print(type(B))
sum = A+B
mul = A*B
div = A/B
A_sqr = A**2
B_sqr = B**2

print(f"the sum of the numbers is {sum}")
print(f"the multiplication of the number is {mul}")
print(f"the division of the number is {div}")
print(f"the square of the number A is {A_sqr}")
print(f"the square of the number B is {B_sqr}")
```
